```python
from PIL import Image

def image_flip(image, horizontal=True, vertical=True):
    if horizontal:
        image = image.transpose(Image.FLIP_LEFT_RIGHT)
    if vertical:
        image = image.transpose(Image.FLIP_TOP_BOTTOM)
    return image
image_path = "C:/Users/vadre/Desktop/IP/Sharbatgula.png"
image = Image.open(image_path)

flipped_image_vert = image_flip(image, horizontal=False, vertical=True)
flipped_image_horiz = image_flip(image, horizontal=True, vertical=False)
horiz_and_vert = image_flip(image, horizontal = True, vertical = True)

display(image, "Original Image")
display(flipped_image_vert, "Flipped Image Vertically")
display(flipped_image_horiz, "Flipped Image Horizontally")
display(horiz_and_vert, "Flipped Image Horizontally and vertically")
```
