```python
from PIL import Image
import numpy as np

original_image = Image.open('C:/Users/vadre/Desktop/IP/Lena.png')
original_image = original_image.convert('RGB')
width, height = original_image.size
pixels = np.array(original_image)

def map_pixels(image, color_table):
    result_image = Image.new("RGB", (width, height))

    for y in range(height):
        for x in range(width):
            pixel = image.getpixel((x, y))
            color_table_array = np.array(color_table)
            closest_color_index = np.argmin(np.linalg.norm(color_table_array - pixel, axis=1))
            closest_color = tuple(color_table[closest_color_index])
            result_image.putpixel((x, y), closest_color)

    return result_image

transformed_image = map_pixels(original_image, universal_color_table)
transformed_image.save('C:/Users/vadre/Desktop/IP/Lena_Trans.png')

```

```python
from PIL import Image
import numpy as np

original = Image.open('C:/Users/vadre/Desktop/IP/Lena.png')
original = original.convert('RGB')

def adaptive_color_table(image, num_colors):
    width, height = image.size
    pixels = np.array(image)
    pixel_count = width * height

    pixels = pixels.reshape((pixel_count, 3))

    color_counts = {}
    for pixel in pixels:
        color_tuple = tuple(pixel)
        if color_tuple in color_counts:
            color_counts[color_tuple] += 1
        else:
            color_counts[color_tuple] = 1

    sorted_colors = sorted(color_counts.keys(), key=lambda x: color_counts[x], reverse=True)
    adaptive_color_table = sorted_colors[:num_colors]

    return adaptive_color_table

def map_to_color_table(image, color_table):
    width, height = image.size
    result_image = Image.new("RGB", (width, height))

    for y in range(height):
        for x in range(width):
            pixel = image.getpixel((x, y))
            closest_color = min(color_table, key=lambda color: np.linalg.norm(np.array(color) - np.array(pixel)))
            result_image.putpixel((x, y), closest_color)

    return result_image

num_colors = 256
adaptive_color = adaptive_color_table(original, num_colors)
transformed_image = map_to_color_table(original, adaptive_color)
transformed_image.save('C:/Users/vadre/Desktop/IP/Lena_Adapt.png')
```
